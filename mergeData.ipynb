```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/makananSertifikasiHalal.csv')
df2 = pd.read_csv('data/MUI.csv')
df3 = pd.read_csv('data/pengesahanProdukMUI.csv')
df4 = pd.read_csv('data/produkHalal.csv')
df5 = pd.read_csv('data/produkHaram.csv')
df6 = pd.read_csv('data/sertifikasiHaram.csv')
df7 = pd.read_csv('data/sertifikasiHalal.csv')
```

```python
# Gabungkan DataFrame menjadi satu
df = pd.concat([df1, df2, df3, df4, df5, df6, df7])
```

```python
# Hapus kolom 'Unnamed: 0' jika ada
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
```

```python
# Reset nomor baris data
df = df.reset_index(drop=True)
```

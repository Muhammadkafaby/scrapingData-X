### Menghapus nilai False

```python
df.info()
```

```python
# Metode 1: Menggunakan drop
df['full_text'] = df['full_text'].astype(str).str.lower()  # Mengubah teks menjadi lowercase
df['full_text'] = df['full_text'].replace('false', pd.NA)  # Mengganti 'false' dengan pd.NA (Not Available)
df = df.dropna(subset=['full_text'])

# Metode 2: Menggunakan boolean indexing
df['full_text'] = df['full_text'].astype(str).str.lower()  # Mengubah teks menjadi lowercase
df = df[df['full_text'] != 'false']
```

```python
df.shape
```

```python
df = df.drop_duplicates(subset=['full_text'])
```

```python
df.duplicated().sum()
```

```python
df = df.dropna(subset=['full_text'])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
def clean_twitter_text(text):
  text = re.sub(r'@[A-Za-z0-9_]+', '', text)
  text = re.sub(r'#\w+', '', text)
  text = re.sub(r'RT[\s]+', '', text)
  text = re.sub(r'https?://\S+', '', text)

  text = re.sub(r'[^A-Za-z0-9 ]', '', text)
  text = re.sub(r'\s+', ' ', text).strip()

  return text

df['full_text'] = df['full_text'].apply(clean_twitter_text)
```
